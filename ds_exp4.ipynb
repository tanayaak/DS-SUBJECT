```python
import pandas as pd

# Load the dataset (adjust the path as needed)
file_path = '/content/sample_data/california_housing_test.csv'  # Replace with your actual file path
df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Normalize numerical columns (excluding non-numeric columns)
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
df['longitude'] = (df['longitude'] - df['longitude'].min()) / (df['longitude'].max() - df['longitude'].min())

# Print the normalized DataFrame
print(df.head())  # Display the first few rows of the normalized DataFrame
```

```python
#BALANCING THE DATASET
```

```python
from imblearn.over_sampling import RandomOverSampler

# Separate features and target variable
X = df.drop('longitude', axis=1)  # Use the correct column name
y = df['longitude']  # Use the correct column name

# Convert target variable to string
y = y.astype(str)

# Encode categorical variables
X = pd.get_dummies(X)

# Initialize and apply the RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)

# Convert resampled data back to DataFrame
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['longitude'] = y_resampled  # Use the correct column name

# Print the balanced dataset
print(df_resampled.head())
```
