```python
import pandas as pd

# Load the dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
wine_data = pd.read_csv(url, sep=';')

# Display the first few rows of the dataset
print(wine_data.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatterplot with hue
sns.scatterplot(x='alcohol', y='quality', hue='pH', data=wine_data, palette='coolwarm')
plt.title('Scatterplot of Alcohol vs. Quality with pH Hue')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.show()
```

```python
# Bar plot for average wine quality by alcohol content
average_quality = wine_data.groupby('alcohol')['quality'].mean().reset_index()

sns.barplot(x='alcohol', y='quality', data=average_quality, palette='Blues')
plt.title('Average Wine Quality by Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Average Quality')
plt.xticks(rotation=90)  # Rotate x labels if there are many categories
plt.show()
```

```python
# Histogram of alcohol content
plt.hist(wine_data['alcohol'], bins=15, color='purple', edgecolor='black')
plt.title('Distribution of Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot of wine quality
sns.boxplot(x='quality', data=wine_data, palette='Set3')
plt.title('Box Plot of Wine Quality')
plt.xlabel('Quality')
plt.show()
```

```python
# Line chart of average pH by alcohol content
average_pH = wine_data.groupby('alcohol')['pH'].mean().reset_index()

plt.plot(average_pH['alcohol'], average_pH['pH'], color='orange', marker='o', linestyle='-', linewidth=2)
plt.title('Average pH by Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Average pH')
plt.grid(True)
plt.show()
```
